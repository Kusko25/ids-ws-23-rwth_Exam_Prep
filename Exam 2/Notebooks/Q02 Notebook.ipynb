```python
## (b)
data_assignment =   [2, 4, 8, 10, 12, 12, 12, 16, 17, 18, 20, 22]
data_odd_case =     [2, 4, 8, 10, 12, 12, 13, 15, 16, 17, 18, 20, 22]
# Definition from 'Instructions\3 - Decision trees and data visualizationexploration\DT Instruction (pen  paper).pdf' Page 6
# Why do you people keep asking stuff like this in exams when clearly you know that
# the definition is ambiguous enough that you have to clarify that 
# "There are multiple methods to compute the quartiles"
# "In this course, we do the following"?
# What does this test, the ability to have memorized a specific method for a problem
# where pretty much all variations of the method produce similarly valid results?

def lecture_defined(data):
    n = len(data)
    if n % 2 == 0:
        left = data[:n//2]
        right = data[n//2:]
        median = (left[-1] + right[0])/2
    else:
        # Obviously we have even length in the assignment but for the sake of completion
        # We'll include the odd length case
        median = data[n//2]
        left = data[:n//2]
        right = data[n//2+1:]

    n_half = len(left) # Both left and right are the same length
    if n_half % 2 == 0:
        # Quartiles are the average of the two middle values
        quart_1 = (left[n_half//2-1] + left[n_half//2])/2
        quart_3 = (right[n_half//2-1] + right[n_half//2])/2
    else:
        # Quartiles are the middle value
        quart_1 = left[n_half//2]
        quart_3 = right[n_half//2]        
    
    # The real MVP here. No wishy washy "depends on the method" here
    mean = sum(data)/n
    return quart_1, median, quart_3, mean

def render(data):
    quart_1, median, quart_3, mean = lecture_defined(data)
    df = {
        "First Quartile": [quart_1,np.quantile(data, 0.25), quantiles(data)[0]],
        "Median": [median, np.median(data), st_median(data)],
        "Third Quartile": [quart_3, np.quantile(data, 0.75), quantiles(data)[2]],
        "Mean": [mean, np.mean(data), st_mean(data)]
    }
    df = pd.DataFrame.from_dict(df, orient='index', columns=['My Answer', 'Numpy', 'Python Statistics'])
    display(df.round())

# Assignment data
print("Assignment Data: ", data_assignment)
render(data_assignment)
# Odd case data
print("Odd Case Data: ", data_odd_case)
render(data_odd_case)
# Comparison Instruction Example
instruction_data = [1, 3, 5, 7, 10, 14, 15, 17, 20, 20, 60]
print("Instruction Data: ", instruction_data)
render(instruction_data)
```
